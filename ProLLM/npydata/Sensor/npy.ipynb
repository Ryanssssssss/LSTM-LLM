```python
import pandas as pd
import numpy as np

# 读取Excel文件
file_path = "/home/wuwujian/LXY/sensor_process/LLM-few/npydata/sensor/mp3b_normal.xlsx"
df = pd.read_excel(file_path, header=None)

# 定义标签映射字典
label_mapping = {
    '乙醇1': 1, '乙醇2': 2, '乙醇3': 3, '乙醇4': 4, '乙醇5': 5, '乙醇6': 6,
    '丙酮1': 7, '丙酮2': 8, '丙酮3': 9, '丙酮4': 10, '丙酮5': 11, '丙酮6': 12,
    '甲醛1': 13, '甲醛2': 14, '甲醛3': 15, '甲醛4': 16, '甲醛5': 17, '甲醛6': 18,
    '甲苯1': 19, '甲苯2': 20, '甲苯3': 21, '甲苯4': 22, '甲苯5': 23, '甲苯6': 24
}

# 获取最后一列的标签
labels = df.iloc[:, -1]

# 使用映射转换标签
y = labels.map(label_mapping)

# 将转换后的标签替换回DataFrame的最后一列
df.iloc[:, -1] = y

# 保存修改后的文件
output_file = "/home/wuwujian/LXY/sensor_process/LLM-few/npydata/sensor/mp3b_normal_up.xlsx"
df.to_excel(output_file, index=False, header=False)

# 打印验证信息
print("原始标签的唯一值：", labels.unique())
print("\n转换后标签的唯一值：", y.unique())
print("\n数据已保存到：", output_file)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# 读取npy文件
base_path = "/home/wuwujian/LXY/sensor_process/TimeDG/npydata/Sensor/"
X_train = np.load(base_path + 'Sensor_train_x.npy')
y_train = np.load(base_path + 'Sensor_train_y.npy')
X_test = np.load(base_path + 'Sensor_test_x.npy')
y_test = np.load(base_path + 'Sensor_test_y.npy')

# 创建图形
plt.figure(figsize=(20, 10))

# 1. 训练集和测试集的标签分布
plt.subplot(2, 2, 1)
train_counts = Counter(y_train)
test_counts = Counter(y_test)

# 获取所有唯一标签
all_labels = sorted(list(set(y_train) | set(y_test)))

# 准备数据
train_dist = [train_counts[label] for label in all_labels]
test_dist = [test_counts[label] for label in all_labels]

x = np.arange(len(all_labels))
width = 0.35

plt.bar(x - width/2, train_dist, width, label='Train')
plt.bar(x + width/2, test_dist, width, label='Test')
plt.xlabel('type')
plt.ylabel('sample_size')
plt.title('Label distribution of training and test sets')
plt.xticks(x, all_labels, rotation=45)
plt.legend()

# 2. 特征值分布（每个通道的平均值分布）
plt.subplot(2, 2, 2)
for i in range(6):
    sns.kdeplot(X_train[:, i, :].mean(axis=1), label=f'Channel {i+1}')
plt.title('Distribution of average feature values ​​of each channel in the training set')
plt.xlabel('Feature average value')
plt.ylabel('Density')
plt.legend()

# 3. 训练集和测试集的特征值范围比较
plt.subplot(2, 2, 3)
train_means = X_train.mean(axis=(0, 2))
test_means = X_test.mean(axis=(0, 2))
channels = [f'Channel {i+1}' for i in range(6)]

x = np.arange(len(channels))
width = 0.35

plt.bar(x - width/2, train_means, width, label='Train')
plt.bar(x + width/2, test_means, width, label='Test')
plt.xlabel('Channel')
plt.ylabel('Average')
plt.title('Comparison of the average values ​​of each channel in the training set and the test set')
plt.xticks(x, channels)
plt.legend()

# 4. 特征值时序分布（每个通道的第一个样本）
plt.subplot(2, 2, 4)
for i in range(6):
    plt.plot(X_train[0, i, :], label=f'Channel {i+1}')
plt.title('Time series feature distribution of the first sample')
plt.xlabel('Time step')
plt.ylabel('Feature value')
plt.legend()

# 调整布局
plt.tight_layout()

# 保存图像
plt.savefig(base_path + 'data_distribution.png', dpi=300, bbox_inches='tight')
plt.close()

# 打印数据集的基本统计信息
print("数据集统计信息：")
print(f"训练集形状: {X_train.shape}")
print(f"测试集形状: {X_test.shape}")
print(f"\n训练集标签数量: {len(y_train)}")
print(f"测试集标签数量: {len(y_test)}")
print(f"\n训练集数值范围: [{X_train.min():.3f}, {X_train.max():.3f}]")
print(f"测试集数值范围: [{X_test.min():.3f}, {X_test.max():.3f}]")
```
